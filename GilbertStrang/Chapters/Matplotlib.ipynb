```````python
# Using numpy array
x = np.array([1,2,3])
y = np.array([4,5,6])

plt.plot(x,y)
plt.show()

# Using y only 
# It assumes that they have equal distance and generate length range index from 0 to len(y)
plt.plot(y)
plt.show()
```````

``````Matplotlib selected x ticks by its own to make graph more readable the actual plotted points are (0,4), (1,5), (2,6). It makes graph readable.

```In only y case``` Matplotlib automatically assigns xticks for best readability the actual points plotted remains same (in this example (0,4), (1,5), (2,6))

```````python
# Decorations

x = np.array([1,2,3])
y = np.array([4,5,6])

plt.plot(x,y)

plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```````

```````python
# adding grid and legend to the graph
x = [0, 1, 2, 3]
y = [0, 1, 4, 9]

plt.plot(x,y, label = "Voltage")
plt.title("Electricity voltage")
plt.xlabel("Time")
plt.ylabel("Volt")
plt.legend()
plt.grid()
plt.show()
```````

## Beyond the Basics

```````python
# Line plot customization 
x = [1,2,3]
y = [2,4,6]

# COLOR (Blue, Green, Red, Cyan, Magenta, Yellow, Black, White)
plt.plot(x, y, color="red")
plt.show()

# LINESTYLE (-,--,:,-.)
plt.plot(x,y, color="Black", linestyle="-.")
plt.show()

# LINEWIDTH (Default 1.5)
plt.plot(x,y, color="green", linestyle="--", linewidth=5)
plt.show()

# MARKER (o,*,s,^,x)
plt.plot(x,y, color="magenta", linestyle=":", linewidth=3, marker="o")
plt.show()
```````

**Scatter plot**

```````python
# Scatter plots
x = [12, -5, 18, 7, -11, 25, -3, 14, 9, -20,
     31, -8, 6, 17, -15, 22, 4, -9, 28, 13,
     -18, 35, 2, -6, 19, 11, -24, 30, 5, -12]

y = [-7, 16, -10, 21, 8, -14, 27, -2, 15, -19,
     4, 24, -13, 9, -28, 18, -5, 32, -11, 6,
     20, -17, 3, 26, -8, 12, -22, 29, -4, 14]

# s is size, marker is shape, alpha is transperancy
plt.scatter(x,y, color="red", marker="^", s=100, alpha=0.25)
plt.title("Scatter plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```````

**Histogram**

```````python
# Histogram
marks = [42, 45, 47, 50, 52, 55, 58,
         61, 63, 65, 66, 68,
         72, 75, 78, 81, 85,
         90, 92, 95]

plt.hist(marks, label="Histogram")
plt.title("Histogram plot")
plt.legend()
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()

# Bins of histogram
plt.hist(marks, bins = 5)
plt.show()
```````

**Bar Charts**

```````python
cat = ["A", "B", "C", "D", "E"]
qua = [10, 20, 30, 40, 50]

plt.bar(cat, qua)
plt.title("Category comparison")
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.show()

plt.bar(cat, qua, label="Cat", color="orange")
plt.title("Category comparison")
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.grid()
plt.legend()
plt.show()
```````

## AI and Plots

```````python
# Old syntax using matrices [plt.subplot(row,column,position)]
x = [1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x,[1,4,9,16])
plt.title("Square")

plt.subplot(2,2,2)
plt.plot(x,[1,8,27,64])
plt.title("Cube")

plt.subplot(2,2,3)
plt.scatter(x,[2,5,3,7])
plt.title("Scatter")

plt.subplot(2,2,4)
plt.bar(["A","B","C"],[3,5,2])
plt.title("Bar")

plt.tight_layout()

plt.show()
```````

```````python
# Modern syntax [(fig,ax = plt.subplot(row,column))]
# fig is the whole figure and ax are the sections of 2x2

x = [1, 2, 3, 4]

# Create a figure with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2, figsize=(8, 6))

# Top-left
ax[0, 0].plot(x, [1, 4, 9, 16])
ax[0, 0].set_title("Square")

# Top-right
ax[0, 1].plot(x, [1, 8, 27, 64])
ax[0, 1].set_title("Cube")

# Bottom-left
ax[1, 0].scatter(x, [2, 5, 3, 7])
ax[1, 0].set_title("Scatter")

# Bottom-right
ax[1, 1].bar(["A", "B", "C"], [3, 5, 2])
ax[1, 1].set_title("Bar")

plt.tight_layout()
plt.show()
```````

**2D matrix as an Image**

```````python
# imshow()
image = np.array([
    [20, 0, 255],
    [180, 100, 255],
    [255, 240, 0]
])

# cmap decides color
plt.imshow(image, cmap="gray")
plt.show()
```````
