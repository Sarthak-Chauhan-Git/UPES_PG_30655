### 📜 Lore Drop: Lists, Tuples & Dictionaries

| Container | Syntax | Mutable? | Ordered? | Use Case |
|-----------|--------|----------|----------|----------|
| **List**  | `[1, 2, 3]` | ✅ Yes | ✅ Yes | Shopping list, scores |
| **Tuple** | `(1, 2, 3)` | ❌ No  | ✅ Yes | Fixed coordinates, RGB |
| **Dict**  | `{"key": val}` | ✅ Yes | ✅ Yes | Spellbook, phone book |

```
LIST  → A bag you can add/remove items from
TUPLE → A sealed artifact (read-only)
DICT  → A map where each location has a name (key)
```


````python
# ── CONTAINER SHOWCASE ────────────────────────────────────────────
# List
inventory = ["sword", "shield", "potion", "potion", "map"]
print("Inventory (List):")
print(" ", inventory)
inventory.append("torch")
print("  After adding torch:", inventory)
inventory.remove("map")
print("  After using map   :", inventory)

# Tuple
coords = (42, 17, 9)
print(f"Location Seal (Tuple): {coords}")

# Dict
spellbook = {"fireball": 50, "ice spike": 35, "heal": 80}
print("Spellbook (Dict):")
for spell, power in spellbook.items():
    print(f"   {spell:12}: {power} power")
````

### 🧪 Quest 1: Frequency of Enemy Appearances

> *In the dungeon log, enemies were spotted multiple times. Count how many times each enemy type appeared.*


````python
enemy_log = ['goblin', 'troll', 'goblin', 'dragon', 'troll',
             'goblin', 'skeleton', 'dragon', 'goblin']

frequency = {}
for enemy in enemy_log:
    frequency[enemy] = frequency.get(enemy, 0) + 1

print("ENEMY SIGHTING REPORT:")
print("─" * 30)
for enemy, count in sorted(frequency.items(), key=lambda x: -x[1]):
    bar = "|" * count
    print(f"  {enemy:12}: {bar} ({count})")
````

### 🧪 Quest 2: The Runner-Up Score

> *The tournament results are in. You need the SECOND highest score — the runner-up.*


````python
scores = [75, 90, 85, 90, 60, 70, 88]

unique_scores = sorted(set(scores), reverse=True)
runner_up = unique_scores[1]

print(f"Tournament scores : {scores}")
print(f"Unique sorted     : {unique_scores}")
print(f"Runner-up score: {runner_up}")
````

### 🧪 Quest 3: The City Spellbook (Dict of Lists)

> *The royal cartographer stores each city's locations in a dictionary. Multiple locations per city.*


````python
city_map = {
    "New York"  : ["Manhattan", "Brooklyn", "Bronx"],
    "California": ["Los Angeles", "San Francisco", "San Diego"],
    "Texas"     : ["Houston", "Dallas", "Austin"]
}

print("ROYAL CITY SPELLBOOK:")
print("─" * 40)
for city, locations in city_map.items():
    print(f"{city}:")
    for loc in locations:
        print(f"└─ {loc}")
````

### ⚔️ Boss Battle: The Movie Vault Filter

> *The vault contains a list of movies with their genres and ratings. A patron demands only Drama films rated above 8.0.*


````python
movies = [
    {"title": "The Shawshank Redemption", "genre": "Drama",  "rating": 9.3},
    {"title": "The Dark Knight",          "genre": "Action", "rating": 9.0},
    {"title": "Forrest Gump",             "genre": "Drama",  "rating": 8.8},
    {"title": "Inception",                "genre": "Sci-Fi", "rating": 8.7},
    {"title": "Schindler's List",         "genre": "Drama",  "rating": 9.0},
    {"title": "The Matrix",               "genre": "Sci-Fi", "rating": 8.7},
    {"title": "Good Will Hunting",        "genre": "Drama",  "rating": 8.3},
]

print("TOP DRAMA FILMS (rating > 8.0):")
print("─" * 45)

filtered = [m for m in movies if m["genre"] == "Drama" and m["rating"] > 8.0]

for m in sorted(filtered, key=lambda x: -x["rating"]):
    print(f" * {m['rating']}  {m['title']}")

print(f"{len(filtered)} films passed the filter. Vault unlocked!")
````
