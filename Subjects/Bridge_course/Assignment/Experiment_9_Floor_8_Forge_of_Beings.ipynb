### 📜 Lore Drop: OOP — The Four Pillars

| Pillar | Meaning | RPG Analogy |
|--------|---------|-------------|
| **Encapsulation** | Bundle data + methods together | A hero's stats are private |
| **Inheritance**   | Child class extends parent | A Paladin IS a Warrior |
| **Polymorphism**  | Same method, different behaviour | All heroes attack differently |
| **Abstraction**   | Hide complexity, show interface | You use `attack()` not the math |

```python
class ClassName:
    def __init__(self, param1, param2):  # Constructor — called on creation
        self.param1 = param1             # Instance attribute

    def method(self):                    # Instance method
        return self.param1
```


````python
# ── FORGING A BASIC HERO ──────────────────────────────────────────
class Hero:
    def __init__(self, name, hp, attack):
        self.name   = name
        self.hp     = hp
        self.attack = attack

    def introduce(self):
        return f"I am {self.name} — HP: {self.hp}, ATK: {self.attack}"

    def take_damage(self, damage):
        self.hp -= damage
        if self.hp <= 0:
            return f"{self.name} has fallen!"
        return f"{self.name} took {damage} damage. HP left: {self.hp}"

# Forge two heroes
hero1 = Hero("Aria", 100, 35)
hero2 = Hero("Zeke", 80, 50)

print(hero1.introduce())
print(hero2.introduce())
print()
print(hero1.take_damage(30))
print(hero2.take_damage(80))
````

### 🧪 Quest 1: The Single Inheritance Chain

> *A Mage is a Hero who also wields spells. They inherit everything from Hero, then add their own powers.*


````python
class Hero:
    def __init__(self, name, hp):
        self.name = name
        self.hp   = hp

    def describe(self):
        return f"Hero: {self.name} | HP: {self.hp}"

class Mage(Hero):
    def __init__(self, name, hp, mana):
        super().__init__(name, hp)  # Call parent constructor
        self.mana = mana

    def cast_spell(self, spell):
        cost = 20
        if self.mana >= cost:
            self.mana -= cost
            return f"{self.name} casts {spell}! (Mana left: {self.mana})"
        return f"Not enough mana!"

    def describe(self):  # Override parent method
        return f"Mage : {self.name} | HP: {self.hp} | Mana: {self.mana}"

mage = Mage("Elara", 70, 150)
print(mage.describe())
print(mage.cast_spell("Fireball"))
print(mage.cast_spell("Ice Storm"))
print(mage.cast_spell("Blizzard"))
print(mage.take_damage(20) if hasattr(mage, 'take_damage') else "No take_damage here")
````

### 🧪 Quest 2: Multilevel Inheritance — The Legendary Line

> *Warrior → Knight → Paladin. Each generation inherits and extends the last.*


````python
class Warrior:
    def __init__(self, name):
        self.name = name
    def fight(self):
        return f"{self.name} swings a sword!"

class Knight(Warrior):
    def __init__(self, name, armor):
        super().__init__(name)
        self.armor = armor
    def defend(self):
        return f"{self.name} raises a {self.armor} shield!"

class Paladin(Knight):
    def __init__(self, name, armor, faith):
        super().__init__(name, armor)
        self.faith = faith
    def holy_strike(self):
        return f"{self.name} channels {self.faith} into a divine blow!"

p = Paladin("Sir Aldric", "Dragonscale", "Lightbringer")
print(p.fight())        # Inherited from Warrior
print(p.defend())       # Inherited from Knight
print(p.holy_strike())  # Paladin's own power
````

### 🧪 Quest 3: Multiple Inheritance — The Hybrid Class

> *Some adventurers are born of two bloodlines. The `HybridWarrior` inherits from both `Swordsman` and `Archer`.*


````python
class Swordsman:
    def melee_attack(self):
        return "Slashes with a blade!"

class Archer:
    def ranged_attack(self):
        return "Fires a precise arrow!"

class HybridWarrior(Swordsman, Archer):
    def __init__(self, name):
        self.name = name
    def combo_attack(self):
        return f"{self.name}: {self.melee_attack()} then {self.ranged_attack()}"

ranger = HybridWarrior("Voss")
print(ranger.melee_attack())
print(ranger.ranged_attack())
print(ranger.combo_attack())
print(f"MRO: {[c.__name__ for c in HybridWarrior.__mro__]}")
````

### ⚔️ Boss: Operator Overloading — The Power Combiner

> *The Forgemaster's final test: two weapons can be combined with `+`. Overload the `__add__` operator to merge their stats.*


````python
class Weapon:
    def __init__(self, name, damage, durability):
        self.name       = name
        self.damage     = damage
        self.durability = durability

    def __str__(self):
        return f"[{self.name} | DMG:{self.damage} | DUR:{self.durability}]"

    def __add__(self, other):
        # Combine two weapons into a legendary one
        combined_name = f"{self.name}-{other.name} Fusion"
        return Weapon(
            combined_name,
            self.damage + other.damage,
            (self.durability + other.durability) // 2
        )

    def __gt__(self, other):
        return self.damage > other.damage

sword  = Weapon("Shadowblade", 80, 100)
staff  = Weapon("Runestaff",   65, 120)
fusion = sword + staff

print(f"Weapon 1  : {sword}")
print(f"Weapon 2  : {staff}")
print(f"Fused     : {fusion}")
print(f"Sword stronger than Staff? {sword > staff}")
print("\nThe Forge of Beings bows to your mastery!")
````
