### 📜 Lore Drop: Conditional Statements

```python
if condition:
    # do this if TRUE
elif other_condition:
    # do this if the first was FALSE but this is TRUE
else:
    # do this if ALL above were FALSE
```

Think of it as:
- `if`   → "Try the first door"
- `elif` → "If that was locked, try this one"
- `else` → "If all else fails, take the escape hatch"


````python
# Quest 1: Is this number worthy? (divisible by both 3 and 5)
num = 45

if num % 3 == 0 and num % 5 == 0:
    print(f"{num} passes the Dual Gate (divisible by 3 AND 5)")
else:
    print(f"{num} is rejected by the Dual Gate")
````

````python
# Quest 2: The Mul-5 Checkpoint
num = 37

if num % 5 == 0:
    print(f"{num} is a multiple of 5 — checkpoint cleared!")
else:
    print(f"{num} is NOT a multiple of 5 — checkpoint locked.")
````

### 🧪 Quest 5: The Quadratic Oracle

> *The Oracle poses an equation: `ax² + bx + c = 0`. She demands to know: are its roots real, or do they exist only in the imaginary realm?*

The **discriminant** `b² - 4ac` reveals the truth:
- `≥ 0` → Real roots (real world)
- `< 0` → Imaginary roots (shadow realm)

`cmath` — the complex math module — handles imaginary numbers.


````python
import cmath

a, b, c = 1, 2, 5  # The Oracle's coefficients

discriminant = b**2 - 4*a*c
root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
root2 = (-b - cmath.sqrt(discriminant)) / (2*a)

if discriminant >= 0:
    print("The roots are REAL — they exist in this world!")
else:
    print("The roots are IMAGINARY — they dwell in the shadow realm!")

print(f"Root 1 : {root1}")
print(f"Root 2 : {root2}")
````

### 🧪 Quest 6: The Leap Year Lock

> *A calendar door. It only opens on **leap years**. The mechanism checks:  
> Divisible by 400 → always open  
> Divisible by 4 but NOT 100 → open  
> Everything else → sealed.*


````python
year = 2024

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(f"{year} is a LEAP YEAR — the calendar door swings open!")
else:
    print(f"{year} is NOT a leap year — the door holds firm.")
````

### 🧪 Quest 7: The Tomorrow Scroll

> *You find a scroll that reveals the next date in the calendar. You must enchant it with any date and it will show what comes after.*


````python
from datetime import date, timedelta

d = date(2005, 9, 20)
next_day = d + timedelta(days=1)

print(f"Today's date  : {d.strftime('%d-%m-%Y')}")
print(f"Tomorrow      : {next_day.strftime('%d-%m-%Y')}")
````

### ⚔️ Boss Battle: The Grade Gate

> *The final door of this floor has seven locks, each labeled with a CGPA range. The Guardian hands you a student's marks. You must calculate their CGPA and determine which lock to open.*


````python
def get_grade(cgpa):
    if 0.0 <= cgpa <= 3.4:   return "F  — The gate is sealed."
    if 3.5 <= cgpa <= 5.0:   return "C+ — A narrow crack opens."
    if 5.1 <= cgpa <= 6.0:   return "B  — A door creaks open."
    if 6.1 <= cgpa <= 7.0:   return "B+ — A proper door opens."
    if 7.1 <= cgpa <= 8.0:   return "A  — The gate swings wide!"
    if 8.1 <= cgpa <= 9.0:   return "A+ — Golden light streams in!"
    if 9.1 <= cgpa <= 10.0:  return "O  — ⭐ Outstanding! The gate EXPLODES open!"
    return "Invalid CGPA"

# Student's subject marks
marks = {"PDS": 70, "Python": 80, "Chemistry": 90, "English": 60, "Physics": 65}

total      = sum(marks.values())
percentage = (total / (len(marks) * 100)) * 100
cgpa       = round(percentage / 10, 1)
grade_msg  = get_grade(cgpa)

print("═══════════════════════════════")
print("    GRADE GATE EVALUATION      ")
print("═══════════════════════════════")
for subject, score in marks.items():
    print(f"  {subject:12}: {score}")
print(f"  Total      : {total}")
print(f"  Percentage : {percentage:.2f}%")
print(f"  CGPA       : {cgpa}")
print(f"  Result     : {grade_msg}")
print("═══════════════════════════════")
````
