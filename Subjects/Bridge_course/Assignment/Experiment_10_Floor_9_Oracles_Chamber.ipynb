## 🔵 ORB 1: NumPy — The Numerical Lens

````python
import numpy as np

# ── 1D ARRAY: The Stat Bar ─────────────────────────────────────────
stats = np.array([45, 78, 92, 60, 55, 88, 71])

print("Hero Combat Stats (NumPy Array):")
print("  Values  :", stats)
print("  Max     :", np.max(stats))
print("  Min     :", np.min(stats))
print("  Mean    :", np.mean(stats))
print("  Std Dev :", np.std(stats).round(2))
print("  Sum     :", np.sum(stats))
````

````python
import numpy as np

# ── 2D ARRAY: The Battle Grid ──────────────────────────────────────
battle_grid = np.array([
    [1, 0, 0, 1],
    [0, 1, 0, 1],
    [1, 1, 0, 0],
    [0, 0, 1, 1]
])

print("Battle Grid (4x4 Matrix):")
print(battle_grid)
print(f"\nGrid shape : {battle_grid.shape}")
print(f"Grid sum   : {battle_grid.sum()} enemies on the map")
print(f"Transposed :\n{battle_grid.T}")
````

````python
import numpy as np

# ── MATRIX OPERATIONS ──────────────────────────────────────────────
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nA + B (Element-wise):")
print(A + B)
print("\nA × B (Matrix Multiplication):")
print(np.dot(A, B))
print("\nDeterminant of A:", np.linalg.det(A))
````

## 🟡 ORB 2: Pandas — The Adventurer's Chronicle

````python
import pandas as pd

# ── CREATE A DATAFRAME: Dungeon Expedition Log ─────────────────────
data = {
    "Adventurer"  : ["Aria", "Zeke", "Voss", "Nira", "Elara"],
    "Class"       : ["Mage", "Warrior", "Archer", "Rogue", "Paladin"],
    "Level"       : [42, 38, 45, 31, 50],
    "HP"          : [380, 520, 310, 270, 490],
    "Kills"       : [120, 98, 145, 67, 200],
    "Gold"        : [4500, 3200, 5800, 2100, 9000]
}

df = pd.DataFrame(data)

print("EXPEDITION CHRONICLE:")
print(df.to_string(index=False))
print(f"\nShape: {df.shape[0]} adventurers × {df.shape[1]} columns")
````

````python
import pandas as pd

data = {
    "Adventurer": ["Aria", "Zeke", "Voss", "Nira", "Elara"],
    "Class"     : ["Mage", "Warrior", "Archer", "Rogue", "Paladin"],
    "Level"     : [42, 38, 45, 31, 50],
    "HP"        : [380, 520, 310, 270, 490],
    "Kills"     : [120, 98, 145, 67, 200],
    "Gold"      : [4500, 3200, 5800, 2100, 9000]
}
df = pd.DataFrame(data)

# Chronicle analysis
print("CHRONICLE STATISTICS:")
print(df.describe().round(1))

print("\nFilter — Adventurers above Level 40:")
print(df[df["Level"] > 40][["Adventurer","Class","Level"]].to_string(index=False))

print("\nSorted by Gold (richest first):")
print(df.sort_values("Gold", ascending=False)[["Adventurer","Gold"]].to_string(index=False))
````

## 🔴 ORB 3: Matplotlib — The Vision of the Oracle

````python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(12, 9))
fig.suptitle("THE ORACLE'S VISION — Floor 9 Dashboard", fontsize=15, fontweight='bold')

adventurers = ["Aria", "Zeke", "Voss", "Nira", "Elara"]
levels = [42, 38, 45, 31, 50]
kills  = [120, 98, 145, 67, 200]
gold   = [4500, 3200, 5800, 2100, 9000]
hp     = [380, 520, 310, 270, 490]
classes = ["Mage", "Warrior", "Archer", "Rogue", "Paladin"]
class_counts = [2, 3, 1, 4, 2]

# ── 1. BAR CHART: Levels ───────────────────────────────────────────
colors = ['#e74c3c','#3498db','#2ecc71','#f39c12','#9b59b6']
axes[0,0].bar(adventurers, levels, color=colors, edgecolor='black')
axes[0,0].set_title("Adventurer Levels")
axes[0,0].set_ylabel("Level")
axes[0,0].set_ylim(0, 60)
for i, (name, lvl) in enumerate(zip(adventurers, levels)):
    axes[0,0].text(i, lvl + 0.5, str(lvl), ha='center', fontweight='bold')

# ── 2. LINE CHART: Gold Earned ─────────────────────────────────────
axes[0,1].plot(adventurers, gold, marker='o', color='gold',
               linewidth=2.5, markersize=10, markerfacecolor='darkgoldenrod')
axes[0,1].fill_between(range(len(adventurers)), gold, alpha=0.15, color='gold')
axes[0,1].set_title("Gold Accumulated")
axes[0,1].set_ylabel("Gold Coins")
axes[0,1].set_xticks(range(len(adventurers)))
axes[0,1].set_xticklabels(adventurers)

# ── 3. SCATTER: HP vs Kills ────────────────────────────────────────
scatter = axes[1,0].scatter(hp, kills, c=levels, cmap='plasma',
                             s=200, edgecolors='black', linewidth=1.5)
plt.colorbar(scatter, ax=axes[1,0], label='Level')
axes[1,0].set_title("HP vs Kills (color = level)")
axes[1,0].set_xlabel("HP")
axes[1,0].set_ylabel("Kills")
for i, name in enumerate(adventurers):
    axes[1,0].annotate(name, (hp[i]+5, kills[i]+2), fontsize=8)

# ── 4. PIE CHART: Class Distribution ──────────────────────────────
explode = [0.05]*5
axes[1,1].pie(class_counts, labels=classes, colors=colors,
              autopct='%1.0f%%', explode=explode, startangle=90)
axes[1,1].set_title("Class Distribution")

plt.tight_layout()
plt.savefig("oracle_vision.png", dpi=120, bbox_inches='tight')
plt.show()
print("\nThe Oracle's Vision has been recorded!")
````
