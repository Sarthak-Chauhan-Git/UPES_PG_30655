### 📜 Lore Drop: Python Operators

| Symbol | Name | RPG Use |
|--------|------|---------|
| `+`  | Addition | Add HP potions |
| `-`  | Subtraction | Take damage |
| `*`  | Multiplication | Multiply attack |
| `/`  | Division | Split loot |
| `//` | Floor Division | Rounds down (integer result) |
| `%`  | Modulo | Leftover after split |
| `**` | Exponentiation | Power-up multiplier |
| `<<` | Left Shift | Multiply by 2ⁿ |
| `>>` | Right Shift | Divide by 2ⁿ |


````python
# ── WEAPON TEST: Basic Arithmetic ─────────────────────────────────
x = 9   # Adventurer attack stat
y = 7   # Enemy defense stat

print("=== COMBAT CALCULATOR ===")
print("ATK + DEF  (Addition)      :", x + y)
print("ATK - DEF  (Subtraction)   :", x - y)
print("ATK * DEF  (Multiplication):", x * y)
print("ATK / DEF  (Division)      :", x / y)
````

### 🧪 Quest 1: The Circular Seal

> *"To unlock the first door," the Blacksmith says, "calculate the area enclosed by this circular rune."*

**Formula:** `area = π × r²`  
The `math` module is your spellbook — it contains pre-built mathematical formulas.


````python
import math

# A circular rune with radius 5.5 units
radius = 5.5
area = math.pi * radius ** 2

print(f"Rune radius : {radius}")
print(f"Sealed area : {area:.4f} square units")
print("The circular seal is broken!")
````

### 🧪 Quest 2: The Pythagorean Bridge

> *A crumbling bridge. Two support beams measure 3 and 4 units. How long is the diagonal support? The bridge won't hold unless you calculate it.*

**Formula:** `c = √(a² + b²)` — Pythagoras' Theorem


````python
import math

a = 3   # Horizontal beam
b = 4   # Vertical beam
c = math.sqrt(a**2 + b**2)

print(f"Beam A      : {a}")
print(f"Beam B      : {b}")
print(f"Diagonal (c): {c}")
print("Bridge repaired!")
````

### 🧪 Quest 3: The Merchant's Interest

> *A shady merchant offers you gold on loan. You must calculate the Simple Interest before signing anything.*

**Formula:** `SI = (P × R × T) / 100`


````python
# Simple Interest Calculator
principal = 10000   # Gold borrowed
rate      = 8       # Interest rate per year
time      = 2       # Years

simple_interest = (principal * rate * time) / 100

print(f"Principal  : {principal} gold")
print(f"Rate       : {rate}%")
print(f"Time       : {time} years")
print(f"Interest   : {simple_interest} gold")
print(f"Total Owed : {principal + simple_interest} gold")
````

### 🧪 Quest 4: The Triangular Terrain

> *You must calculate the area of a triangular battlefield using only the three side lengths.*

**Heron's Formula:** `s = (a+b+c)/2`,  `area = √(s(s-a)(s-b)(s-c))`


````python
import math

a, b, c = 3, 4, 5          # Three sides of triangle
s = (a + b + c) / 2        # Semi-perimeter
area = math.sqrt(s * (s-a) * (s-b) * (s-c))

print(f"Sides       : {a}, {b}, {c}")
print(f"Semi-perim  : {s}")
print(f"Battlefield area: {area:.2f} sq units")
````

### 🧪 Quest 5: The Time Crystal

> *A time crystal stores 7384 seconds of recorded memory. Decode it into hours, minutes, and seconds.*


````python
total_seconds = 7384

hours   = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60

print(f"Time Crystal encodes: {total_seconds} seconds")
print(f"Decoded → {hours}h {minutes}m {seconds}s")
````

### 🔧 The Bit-Forge: Bitwise & Shift Operators

> *Deep in the Armory is a special weapon rack: the Bit-Forge. These weapons work at the binary level — the language of machines.*

**Bitwise operators** treat numbers as binary (0s and 1s):
- `&` AND — both bits must be 1
- `|` OR  — either bit can be 1  
- `^` XOR — bits must be different
- `<<` Left shift — multiply by 2ⁿ
- `>>` Right shift — divide by 2ⁿ


````python
# Truth table for bitwise weapons
print("┌───┬───┬─────┬─────┬─────┐")
print("│ a │ b │ a&b │ a|b │ a^b │")
print("├───┼───┼─────┼─────┼─────┤")
for a in (0, 1):
    for b in (0, 1):
        print(f"│ {a} │ {b} │  {a&b}  │  {a|b}  │  {a^b}  │")
print("└───┴───┴─────┴─────┴─────┘")
````

````python
# Shift Cannons
num   = 10
shift = 2

print(f"Number           : {num}  (binary: {bin(num)})")
print(f"Left Shift  << 2 : {num << shift}  (binary: {bin(num << shift)})")
print(f"Right Shift >> 2 : {num >> shift}  (binary: {bin(num >> shift)})")
````

````python
# Membership Operator — is this number in the enemy's sequence?
sequence = (10, 20, 56, 78, 89)
number   = 56

print(f"Enemy sequence : {sequence}")
print(f"Searching for  : {number}")
print(f"Found?         : {number in sequence}")

# Is this character a rune-letter?
text = "Python Programming"
ch   = "g"
print(f"\nRune '{ch}' in '{text}'? → {ch in text}")
````
