### 📜 Lore Drop: Strings & Sets

**Strings** are immutable sequences of characters:
```python
s = "Python"
s[0]      # 'P'      — indexing
s[1:4]    # 'yth'    — slicing
s[::-1]   # 'nohtyP' — reverse
len(s)    # 6        — length
```

**Sets** store unique items only:
```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
A | B   # Union        {1,2,3,4,5,6}
A & B   # Intersection {3,4}
A - B   # Difference   {1,2}
```


````python
# ── STRING TOOLS DEMO ─────────────────────────────────────────────
s = "Python Programming"

print("Original      :", s)
print("Uppercase     :", s.upper())
print("Lowercase     :", s.lower())
print("Length        :", len(s))
print("Reversed      :", s[::-1])
print("First 6 chars :", s[:6])
print("Replace       :", s.replace("Python", "Dragon"))
print("Split words   :", s.split())
````

### 🧪 Quest 1: The Vowel Counter

> *The vault door counts vowels in the password phrase. Wrong count = sealed forever.*


````python
sentence = "The quick brown fox jumps over the lazy dog"
vowels = "aeiouAEIOU"

count = sum(1 for ch in sentence if ch in vowels)
print(f"Phrase  : {sentence}")
print(f"Vowels  : {count}")
print("Vault door opens!" if count > 0 else "Silence.")
````

---
### 🧪 Quest 2: The Capital Rune Counter

> *An ancient inscription requires you to count how many Capital Runes (uppercase letters) appear.*


````python
text = "Hello World this Is A Python Program"

capital_count = sum(1 for ch in text if ch.isupper())
print(f"Text            : {text}")
print(f"Capital Runes   : {capital_count}")
````

---
### 🧪 Quest 3: The Substring Cipher

> *A message has been encoded. Find how many times a keyword appears in the message — that's the key to the cipher lock.*


````python
main_string = "the cat sat on the mat near the vat"
sub_string  = "the"

count = main_string.count(sub_string)
print(f"Message     : {main_string}")
print(f"Keyword     : '{sub_string}'")
print(f"Occurrences : {count} — cipher key found!")
````

### 🧪 Quest 4: The Reverse Echo Chamber

> *Speak a word into the chamber. It echoes back in reverse. Is it the same? Then you've found a STRING PALINDROME.*


````python
word = "racecar"

if word == word[::-1]:
    print(f"'{word}' echoes the same — it's a palindrome!")
else:
    print(f"'{word}' echoes differently — not a palindrome.")
````

### 🧪 Quest 5: The Set Circles

> *Three glowing circles on the floor: Circle A, Circle B. Where they overlap is sacred ground. Where they don't — contested territory.*


````python
A = {10, 20, 30, 40, 50}
B = {30, 40, 50, 60, 70}

print(f"Circle A          : {sorted(A)}")
print(f"Circle B          : {sorted(B)}")
print(f"Union    (A ∪ B)  : {sorted(A | B)}")
print(f"Intersect(A ∩ B)  : {sorted(A & B)}")
print(f"Diff     (A − B)  : {sorted(A - B)}")
print(f"Sym Diff (A △ B)  : {sorted(A ^ B)}")
````

### ⚔️ Boss: The Anagram Oracle

> *Two words appear on separate stone tablets. The Oracle demands to know: are they anagrams of each other?  
> (Anagram = same letters, different order)*


````python
def are_anagrams(word1, word2):
    return sorted(word1.lower()) == sorted(word2.lower())

pairs = [("listen", "silent"), ("hello", "world"), ("Triangle", "Integral")]

print("ANAGRAM ORACLE RESULTS:")
print("─" * 40)
for w1, w2 in pairs:
    result = "ANAGRAM" if are_anagrams(w1, w2) else "NOT anagram"
    print(f"  '{w1}' & '{w2}' → {result}")
````
