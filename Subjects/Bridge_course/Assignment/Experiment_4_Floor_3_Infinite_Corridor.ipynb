### 📜 Lore Drop: Python Loops

Two loop weapons exist in your arsenal:

**`for` loop** — Runs a fixed number of times (you know in advance)
```python
for i in range(5):       # runs 5 times: i = 0,1,2,3,4
    print(i)
```

**`while` loop** — Runs as long as a condition is True (may never end!)
```python
while hp > 0:
    hp -= 10             # keep going until HP hits zero
```

**Escape tools:**
| Keyword | Effect |
|---------|--------|
| `break`    | Exit the loop immediately |
| `continue` | Skip this iteration, go to next |
| `pass`     | Do nothing (placeholder) |


````python
# ── THE CORRIDOR MAP: for vs while ────────────────────────────────
print("=== FOR LOOP: Counting torches ===")
for i in range(1, 6):
    print(f"  Torch #{i}")

print()
print("=== WHILE LOOP: Walking until HP runs out ===")
hp = 50
steps = 0
while hp > 0:
    hp -= 10
    steps += 1
    print(f"  Step {steps} — HP remaining: {hp}")
print("You collapsed. The corridor wins... for now.")
````

### 🧪 Quest 1: The Factorial Scroll

> *A sealed door bears the inscription: `n!`  
> To open it, multiply every integer from 1 to n.*

`5! = 5 × 4 × 3 × 2 × 1 = 120`


````python
# Factorial using a loop
n = 5
factorial = 1

for i in range(1, n + 1):
    factorial *= i

print(f"{n}! = {factorial}")
print("The Factorial Scroll door opens!")
````

---
### 🧪 Quest 2: The Fibonacci Stairs

> *Each stair is the sum of the two before it. A strange enchantment — but once you see the pattern, you can climb forever.*

`0, 1, 1, 2, 3, 5, 8, 13, 21 ...`


````python
# Fibonacci sequence — first 10 stairs
n = 10
a, b = 0, 1

print("Fibonacci Stairs:")
print(a, b, end=" ")

for _ in range(n - 2):
    a, b = b, a + b
    print(b, end=" ")
print()
print("You've climbed the Fibonacci Stairs!")
````

### 🧪 Quest 3: The Armstrong Chamber

> *A room only opens for Armstrong numbers — those whose digits, cubed and summed, equal the number itself.*
>
> `153 → 1³ + 5³ + 3³ = 1 + 125 + 27 = 153` ✅ Armstrong!


````python
# Find all Armstrong numbers between 1 and 500
print("Armstrong Numbers (1–500):")
armstrong_found = []

for num in range(1, 501):
    digits = str(num)
    n = len(digits)
    total = sum(int(d)**n for d in digits)
    if total == num:
        armstrong_found.append(num)

print(armstrong_found)
print(f"Total found: {len(armstrong_found)} Armstrong rooms unlocked!")
````

---
### 🧪 Quest 4: The Prime Sentinels

> *Prime numbers are the rarest guards in the corridor — divisible by only themselves and 1. Find all of them up to 50.*


````python
# Sieve of Primes up to 50
print("Prime Sentinels (up to 50):")
primes = []

for num in range(2, 51):
    is_prime = True
    for divisor in range(2, int(num**0.5) + 1):
        if num % divisor == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)

print(primes)
print(f"Total sentinels: {len(primes)}")
````

---
### 🧪 Quest 5: The Palindrome Mirror

> *A mirror in the corridor shows your number in reverse. If the reverse looks the same — it's a palindrome. The mirror door only opens for palindromes.*


````python
# Check if a number is a palindrome
num = 12321

reversed_num = int(str(num)[::-1])

if num == reversed_num:
    print(f"{num} is a PALINDROME — the mirror door opens!")
else:
    print(f"{num} is NOT a palindrome — the mirror cracks.")
````

### ⚔️ Boss Battle: The Multiplication Labyrinth

> *The corridor's final challenge — a 10×10 multiplication maze. Every cell must be filled to escape.*


````python
# Multiplication table maze (1 to 5, printed neatly)
print("MULTIPLICATION LABYRINTH")
print("    ", end="")
for j in range(1, 11):
    print(f"{j:4}", end="")
print()
print("   " + "─"*41)

for i in range(1, 6):
    print(f"{i:2} │", end="")
    for j in range(1, 11):
        print(f"{i*j:4}", end="")
    print()

print()
print("You navigated the labyrinth!")
````
