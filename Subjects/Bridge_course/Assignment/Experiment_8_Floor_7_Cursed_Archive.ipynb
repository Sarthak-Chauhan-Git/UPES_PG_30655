### 📜 Lore Drop: Exception Handling & File I/O

```python
try:
    # Attempt the dangerous action
    result = risky_operation()
except SomeError as e:
    # Handle the specific failure
    print("Error caught:", e)
else:
    # Only runs if NO exception occurred
    print("All clear!")
finally:
    # ALWAYS runs, no matter what
    print("Cleanup done.")
```

**File operations:**
```python
with open("file.txt", "w") as f:   # 'w'=write, 'r'=read, 'a'=append
    f.write("Hello Archive")
```
The `with` keyword auto-closes the file — like a self-sealing vault.


````python
# ── WRITING TO THE ARCHIVE ────────────────────────────────────────
filename = "dungeon_log.txt"

with open(filename, "w") as f:
    f.write("Day 1: Entered the tower.\n")
    f.write("Day 2: Cleared Floor 3 — the Infinite Corridor.\n")
    f.write("Day 3: Defeated the Fibonacci Golem.\n")
    f.write("Day 4: Reached the Cursed Archive.\n")

print(f"Dungeon log written to '{filename}'")
````

````python
# ── READING THE ARCHIVE ───────────────────────────────────────────
with open("dungeon_log.txt", "r") as f:
    contents = f.read()

print("DUNGEON LOG:")
print("─" * 40)
print(contents)
````

````python
# ── APPENDING A NEW ENTRY ─────────────────────────────────────────
with open("dungeon_log.txt", "a") as f:
    f.write("Day 5: Learned file handling. The Archive bows.\n")

with open("dungeon_log.txt", "r") as f:
    lines = f.readlines()

print(f"Updated log ({len(lines)} entries):")
for i, line in enumerate(lines, 1):
    print(f"  {i}. {line}", end="")
````

### 🧪 Quest 1: The Missing Scroll (FileNotFoundError)

> *You reach for a scroll that doesn't exist. Without protection, the curse strikes immediately. With `try-except`, you survive.*


````python
# Without protection — would crash:
# open("missing_scroll.txt", "r")  ← DANGER

# With armour:
try:
    with open("missing_scroll.txt", "r") as f:
        data = f.read()
        print(data)
except FileNotFoundError:
    print("Scroll not found! The shelf is empty.")
    print("Creating a new one for future use...")
    with open("missing_scroll.txt", "w") as f:
        f.write("This scroll was created as a replacement.")
    print("Replacement scroll created.")
````

### 🧪 Quest 2: The Division Demon (ZeroDivisionError)

> *Dividing anything by zero tears a hole in reality. The Demon climbs out. Handle it before it does.*


````python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        return "Division Demon blocked! Cannot divide by zero."
    else:
        return f"{a} / {b} = {result}"
    finally:
        print("[Division attempt complete]")

print(safe_divide(10, 2))
print(safe_divide(7, 0))
print(safe_divide(100, 4))
````

### 🧪 Quest 3: The Input Curse (ValueError & TypeError)

> *The Archive accepts only numbers at its input gate. A traveller types "sword" instead of a number — the gate rejects them.*


````python
def get_level(user_input):
    try:
        level = int(user_input)
        if level < 1:
            raise ValueError("Level must be at least 1!")
        return f"Hero registered at Level {level}"
    except ValueError as e:
        return f"Invalid level input: {e}"

test_inputs = ["42", "sword", "-5", "0", "100"]
print("LEVEL REGISTRATION GATE:")
print("─" * 40)
for inp in test_inputs:
    print(f"  Input '{inp:6}' → {get_level(inp)}")
````
