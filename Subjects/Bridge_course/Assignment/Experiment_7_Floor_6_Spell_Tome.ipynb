### 📜 Lore Drop: Functions

```python
def spell_name(ingredient1, ingredient2):
    result = ingredient1 + ingredient2   # the magic happens here
    return result                        # send the result back

# Call the spell:
output = spell_name(10, 20)
```

**Types of spells:**
| Type | Description |
|------|-------------|
| **Default args** | `def greet(name="Adventurer")` |
| **`*args`**      | Accept any number of positional args |
| **`**kwargs`**   | Accept any number of keyword args |
| **Lambda**       | A one-line anonymous spell |
| **Recursive**    | A spell that calls itself |


````python
# ── BASIC SPELL DEFINITION ────────────────────────────────────────
def greet(name="Adventurer"):
    return f"Hail, {name}! Welcome to Floor 6."

print(greet())
print(greet("Sarthak"))
````

---
### 🧪 Quest 1: Max & Min Without Magic Items

> *The Tome Keeper confiscates your `max()` and `min()` built-in runes. You must find the greatest and smallest values using only your own logic.*


````python
def find_max(lst):
    maximum = lst[0]
    for num in lst[1:]:
        if num > maximum:
            maximum = num
    return maximum

def find_min(lst):
    minimum = lst[0]
    for num in lst[1:]:
        if num < minimum:
            minimum = num
    return minimum

numbers = [34, 12, 78, 5, 91, 23, 67]

print(f"Number pool  : {numbers}")
print(f"Maximum      : {find_max(numbers)} ")
print(f"Minimum      : {find_min(numbers)} ")
````

---
### 🧪 Quest 2: The Recursive Staircase

> *Some spells invoke themselves. This is called recursion — a spell that casts itself until the answer reveals itself.*
>
> `factorial(5) → 5 × factorial(4) → ... → 5 × 4 × 3 × 2 × 1`


````python
def factorial(n):
    if n == 0 or n == 1:    # Base case — the recursion stops here
        return 1
    return n * factorial(n - 1)   # Recursive call

for i in range(1, 8):
    print(f"  {i}! = {factorial(i)}")
````

---
### 🧪 Quest 3: The Infinite Ingredient Cauldron (*args)

> *Some spells don't know how many ingredients they'll receive. `*args` accepts any number.*


````python
def sum_all(*args):
    # Sum any number of ingredients
    total = 0
    for num in args:
        total += num
    return total

print("Cauldron mixing:")
print("  3 ingredients     :", sum_all(10, 20, 30))
print("  5 ingredients     :", sum_all(1, 2, 3, 4, 5))
print("  7 ingredients     :", sum_all(5, 10, 15, 20, 25, 30, 35))
````

### 🧪 Quest 4: The Named Rune Chest (**kwargs)

> *`**kwargs` lets you pass named arguments — like labelled potion vials — into any spell.*


````python
def display_hero(**kwargs):
    # Display an adventurer profile from keyword arguments
    print("ADVENTURER PROFILE:")
    print("─" * 30)
    for key, value in kwargs.items():
        print(f"  {key:12}: {value}")

display_hero(name="Aria", class_type="Mage", level=42, hp=380, mp=920)
````

---
### 🧪 Quest 5: Lambda — The Instant Enchantment

> *Sometimes you need a quick one-line spell with no name. That's a lambda — a throwaway enchantment.*


````python
# Lambda spells
square    = lambda x: x ** 2
cube      = lambda x: x ** 3
add_two   = lambda x, y: x + y
is_even   = lambda x: "Even" if x % 2 == 0 else "Odd"

print("Lambda Enchantments:")
print(f"  square(7)   = {square(7)}")
print(f"  cube(4)     = {cube(4)}")
print(f"  add(3,9)    = {add_two(3,9)}")
print(f"  is_even(14) = {is_even(14)}")

# Lambda with sorted()
heroes = [("Aria", 42), ("Zeke", 31), ("Voss", 57), ("Nira", 25)]
ranked = sorted(heroes, key=lambda h: h[1], reverse=True)
print(" Heroes ranked by level:")
for name, level in ranked:
    print(f"   {name:8} — Level {level}")
````

### ⚔️ Boss Battle: The Fibonacci Recursion Golem

> *The final guardian of this floor is the Fibonacci Golem — it only falls if you can summon any Fibonacci number recursively.*


````python
def fibonacci(n):
    if n <= 0: return 0
    if n == 1: return 1
    return fibonacci(n-1) + fibonacci(n-2)

print("FIBONACCI GOLEM — Striking at each step:")
for i in range(1, 11):
    val = fibonacci(i)
    bar = "|" * val if val < 40 else "|" * 40 + "..."
    print(f"  fib({i:2}) = {val:4}  {bar}")

print("The Golem dissolves into a sequence.")
````
