### 📜 Lore Drop: What is Python?

Python is a **high-level, interpreted programming language** — meaning the Interpreter reads your code line by line and executes it immediately. No compiling. No delays. Speak, and the world responds.

There are two ways to run Python:
| Mode | What it does |
|---|---|
| **Interactive Mode** | Run one line at a time (like talking to the Interpreter directly) |
| **Script Mode** | Write a full `.py` file and run it all at once (like casting a prepared spell scroll) |


````python
import sys

# The Interpreter awakens and introduces itself
print("THE INTERPRETER SPEAKS:")
print("Python Version:", sys.version)
print("The dungeon is ready. Your adventure begins now.")
````

### 🧙 Step 1: Define Your Character

> *The runes on the wall pulse. "Declare yourself," the voice commands. "Give your variables meaning."*

In Python, a **variable** is a name that stores a value. You don't need to declare a type — Python figures it out.

```
  [INTEGER]   A whole number         → health = 100
  [FLOAT]     A decimal number       → speed = 3.14
  [STRING]    Text / words           → name = "Aria"
  [BOOLEAN]   True or False          → is_alive = True
  [LIST]      A collection of items  → inventory = [1, 2, 3]
```


````python
# ── CHARACTER SHEET ──────────────────────────────────────────────
integer_value = 10          # Health Points (whole number)
float_value   = 3.14        # Speed stat   (decimal)
string_value  = "Python"    # Your weapon of choice
boolean_value = True        # Alive? Yes.
list_value    = [1, 2, 3]   # Starter inventory

print("=== CHARACTER STATS ===")
print("Health Points (int)  :", integer_value)
print("Speed Stat    (float):", float_value)
print("Weapon        (str)  :", string_value)
print("Alive?        (bool) :", boolean_value)
print("Inventory     (list) :", list_value)
````


---
### 🧙 Step 2: Name Your Hero

> *"A hero without a name is forgotten," the Interpreter warns. "Combine your names. State yourself fully."*

Strings can be **concatenated** (joined together) using `+`. The f-string `f"..."` lets you embed variables directly inside text using `{curly braces}`.


````python
# Joining strings — the adventurer introduces themselves
first = "Sarthak"
last  = "Chauhan"
full_name = first + " " + last
print("Full Name:", full_name)

# f-strings: the mage's shorthand
nickname   = "The Debugger"
first_name = "George"
last_name  = "Washington"
print(f"{first_name} ({nickname}) {last_name}")
````
